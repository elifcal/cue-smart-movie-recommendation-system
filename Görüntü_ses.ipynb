```python
#Fragmandan sesi indirme
import yt_dlp
import imageio_ffmpeg

ffmpeg_yolu = imageio_ffmpeg.get_ffmpeg_exe()

def ses_indir(url, cikti_klasor):
    secenekler = {
        'format': 'bestaudio/best',
        'outtmpl': f'{cikti_klasor}/%(title)s.%(ext)s',
        'postprocessors': [{
            'key': 'FFmpegExtractAudio',
            'preferredcodec': 'mp3',
            'preferredquality': '128',
        }],
        'ffmpeg_location': ffmpeg_yolu,  # ← BUNU EKLE
        'quiet': False,
    }

    with yt_dlp.YoutubeDL(secenekler) as ydl:
        bilgi = ydl.extract_info(url, download=True)
        baslik = bilgi.get('title', 'bilinmiyor')
        sure = bilgi.get('duration', 0)
        print(f"\n✅ İndirme tamamlandı!")
        print(f"   Film: {baslik}")
        print(f"   Süre: {sure} saniye")
        return baslik

baslik = ses_indir(TEST_URL, CIKTI_KLASOR)
```

```python
import librosa
import numpy as np

# İndirilen ilk MP3'ü bul
mp3_dosyasi = None
for d in os.listdir(CIKTI_KLASOR):
    if d.endswith(".mp3"):
        mp3_dosyasi = os.path.join(CIKTI_KLASOR, d)
        break

if mp3_dosyasi:
    print(f"🎵 Yükleniyor: {mp3_dosyasi}")
    y, sr = librosa.load(mp3_dosyasi, duration=60)  # İlk 60 saniye yeter

    sure = librosa.get_duration(y=y, sr=sr)
    print(f"\n📊 Ses Bilgileri:")
    print(f"   Örnekleme hızı : {sr} Hz")
    print(f"   Yüklenen süre  : {sure:.1f} saniye")
    print(f"   Örnek sayısı   : {len(y)}")
    print(f"\n✅ Gün 1 tamamlandı! librosa sesi başarıyla okudu.")
else:
    print("❌ MP3 dosyası bulunamadı, Hücre 4'ü tekrar kontrol et")
```

```python
import imageio_ffmpeg
import os
import shutil

# Tam binary yolunu gör
ffmpeg_exe = imageio_ffmpeg.get_ffmpeg_exe()
print(f"Binary yolu: {ffmpeg_exe}")
print(f"Dosya var mı: {os.path.exists(ffmpeg_exe)}")

# ffmpeg adıyla sembolik link oluştur
hedef = os.path.join(os.path.dirname(ffmpeg_exe), "ffmpeg")
if not os.path.exists(hedef):
    shutil.copy(ffmpeg_exe, hedef)
    print(f"✅ Kopyalandı: {hedef}")
else:
    print(f"✅ Zaten var: {hedef}")

# PATH'e ekle
os.environ["PATH"] = os.path.dirname(ffmpeg_exe) + os.pathsep + os.environ["PATH"]

# Test
import subprocess
r = subprocess.run(["ffmpeg", "-version"], capture_output=True, text=True)
print("✅ ffmpeg çalışıyor!" if r.returncode == 0 else f"❌ Hata: {r.stderr[:100]}")
```

```python
import whisper
import os
import imageio_ffmpeg

# imageio_ffmpeg'in binary yolunu al ve PATH'e ekle
ffmpeg_dir = os.path.dirname(imageio_ffmpeg.get_ffmpeg_exe())
os.environ["PATH"] = ffmpeg_dir + os.pathsep + os.environ["PATH"]

# Kontrol et
import subprocess
result = subprocess.run(["ffmpeg", "-version"], capture_output=True, text=True)
if result.returncode == 0:
    print("✅ ffmpeg PATH'te bulundu")
else:
    print("❌ Hâlâ bulunamadı")

# Şimdi Whisper'ı çalıştır
model = whisper.load_model("base")

ses_dosyasi = "./test_audio/Inception (2010) Official Trailer #1 - Christopher Nolan Movie HD.mp3"

print("⏳ Transkript çıkarılıyor...")
sonuc = model.transcribe(ses_dosyasi)

print("\n📝 Transkript (ilk 500 karakter):")
print(sonuc["text"][:500])
print("\n✅ Whisper başarıyla çalıştı!")
```

```python
subprocess.run([sys.executable, "-m", "pip", "install", "colorthief", "requests"], check=True)

from colorthief import ColorThief
import requests
from io import BytesIO

# Inception posteri (TMDB'den)
poster_url = "https://image.tmdb.org/t/p/w500/9gk7adHYeDvHkCSEqAvQNLV5Uge.jpg"
response = requests.get(poster_url)
ct = ColorThief(BytesIO(response.content))

dominant = ct.get_color(quality=1)
palette = ct.get_palette(color_count=5)

print(f"✅ Dominant renk (RGB): {dominant}")
print(f"✅ 5 renk paleti:")
for i, renk in enumerate(palette):
    print(f"   Renk {i+1}: RGB{renk}")
```
